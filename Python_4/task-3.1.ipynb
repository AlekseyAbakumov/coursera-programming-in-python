### *РЕШЕНИЕ*

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

```python
data = df.iloc[:, :100]
target = df.iloc[:, 100:]
```

```python
train_data, test_data = train_test_split(data, random_state=42, test_size=0.33)
train_target, test_target = train_test_split(target, random_state=42, test_size=0.33)
```

```python
model = Ridge(random_state=42)
```

```python
model.fit(train_data, train_target)
```

```python
predict = model.predict(test_data)
```

```python
error = mean_squared_error(predict, test_target)
error
```

```python
model.score(test_data, test_target)
```

```python
model = Ridge(random_state=42)
```

```python
model.fit(train_data, train_target)
```

```python
predict = model.predict(test_data)
```

```python
error = mean_squared_error(predict, test_target)
error
```

```python
model.score(test_data, test_target)
```
