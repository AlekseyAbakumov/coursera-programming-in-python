```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/Users/ivan/Desktop/fifa-18-demo-player-dataset/df.csv')
```

```python
bins=pd.cut(df['Overall'],np.arange(0,150,2))
```

```python
meanvalue=df.groupby(bins)['Overall'].agg(['count','mean'])
```

```python
graph=plt.scatter(x=meanvalue['count'],y=meanvalue['mean'],c='red')
plt.xlabel('Amount of players with corresponding average overall value')
plt.ylabel('Average overall of the player in a bin')
plt.title('Matplotlib Graph 1 -  amount of players on average overall value in a Bin')
print('Central limit theorem in the dataset')
```
