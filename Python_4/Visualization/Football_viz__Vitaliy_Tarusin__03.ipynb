# Сравнительные гистограммы уровня игрока и его потенциала

Гистограммы позволяют проследить распределение наилучших и наиболее перспективных игроков.
Гистограммы показывают, что у лучших футболистов серьезная конкуренция, т.к. гораздо большее кол-во игроков обладает высоким потенциалом

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'fifa_data\CompleteDataset.csv', low_memory=False)
del df['Flag']
del df['Club Logo']
del df['Photo']
```

```python
fig,ax = plt.subplots(figsize=(20,15))
ax.hist(df.Overall, label='Overall', histtype='step')
ax.hist(df.Potential, label='Potential', histtype='step')
ax.set_title('Распределение по рейтингу', fontdict={'fontsize':24})
plt.ylabel('Кол-во игроков')
plt.xlabel('Рейтинг')
plt.legend()
```
