```python
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
df=pd.read_csv('/Users/ivan/Desktop/fifa-18-demo-player-dataset/df.csv')
```

```python
spanish=df[df['Nationality']=='Spain']
```

```python
spanishclubs=spanish.groupby(['Club'])['Overall'].agg(['max','count'])
```

```python
layout=go.Layout(xaxis=dict(title='Max performance of a spanish player in a certain club'),yaxis=dict(title='Amount of spanish players in a club'))
```

```python
iplot([go.Histogram2dContour(x=spanishclubs['max'], y=spanishclubs['count'], contours=dict(coloring='heatmap')),
       go.Scatter(x=spanishclubs['max'], y=spanishclubs['count'], mode='markers', marker=dict(color='white', size=3, opacity=0.3))], show_link=False)
print('Filament in amount of spanish players in a club - normal distibution')
```
