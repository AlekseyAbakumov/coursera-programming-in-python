```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=150, n_features=2, random_state=1)
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, random_state=1, init='k-means++', n_init=10)
```

```python
kmeans.fit(X)
```

```python
labels = kmeans.labels_
```

```python
labels
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels)
```

```python
crit = []

for k in range(2, 8):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X)
    
    crit.append(kmeans.inertia_)
```

```python
crit
```

```python
plt.plot(range(2, 8), crit)
```

```python
kmeans = KMeans(n_clusters=3, random_state=1, init='k-means++', n_init=10)

kmeans.fit(X)

labels = kmeans.labels_

plt.scatter(X[:, 0], X[:, 1], c=labels)
```
